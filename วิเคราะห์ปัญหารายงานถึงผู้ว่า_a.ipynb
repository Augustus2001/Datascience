```python
data['น้ำ'] = (data['comment'].str.find('น้ำ')!=-1).astype('int')
```

```python
pdic = {}
pdic['น้ำท่วม'] = ['น้ำท่วม','น้ำขัง','ระบายน้ำ','ฝนตก']
pdic['มืด'] = ['มืด','สว่าง','เปลี่ยว']
pdic['จราจร'] = ['รถติด','จราจร']
pdic['กลิ่น'] = ['สกปรก','ขยะ','เหม็น','กลิ่น','น้ำเสีย']
pdic['สัตว์'] = ['แมว','สุนัข','จรจัด','สัตว์']
pdic['ทางเดิน'] = ['ทางเดิน','ทางเท้า','ฟุตบาท','กีดขวาง','ทางข้าม','จอดรถ']
pdic['ถนน'] = ['หลุม','ป้ายบอกทาง','ถนนขรุขระ']
pdic['ไม่ปลอดภัย'] = ['ไม่ปลอดภัย','อันตราย']
pdic["สวน"]=["สวน","พื้นที่สีเขียว","ทางจักรยาน","วิ่ง"]
pdic["สุขภาพจิต"]=["สุขภาพจิต","เศร้า","เครียด","เหงา"]
pdic["สุขภาพกาย"]=["โรค","ฝุ่น","ออกกำลัง"]
```

```python
import numpy as np
def createfeat(data):
  ndata = data.copy()
  for k in pdic.keys():
    wList = pdic[k]
    dList = []
    for w in wList:
      dList.append(data['comment'].str.find(w)!=-1)
    dList = (np.vstack(dList).sum(0)>0).astype('int')
    ndata[k] = dList
  return ndata
```

```python
def clean_feat(feat):
  res = feat.describe()
  res = res.T
  res = res.reset_index()
  res = res[['index','mean']]
  res.columns = ['problem','percent']
  res['percent'] = res['percent']*100
  res=res.drop([0],axis=0)
  res = res.sort_values(by='percent',ascending=False)
  n=len(feat)
  return res,n
```

```python

a=feat.loc[feat["district"].isin(["ทวีวัฒนา","คันนายาว","ปทุมวัน","บางชุนเทียน","หลักสี่"])]
res,n=clean_feat(a)
res=res.sort_values(by="problem")
print(res)
print(res.plot.bar(x='problem',grid=True,title="top_5_parks จำนวน:"+str(n),figsize = (10,5)).get_figure().savefig("n_30_all.png"))
```

```python
res_all,n_all=clean_feat(feat)
res_all=res_all.sort_values(by="problem")
print(res_all)
print(res_all.plot.bar(x='problem',grid=True,title="จำนวน:"+str(n_all),figsize = (10,5)).get_figure().savefig('output.png'))
```

```python
top5_m=res.iloc[:,1].astype(int)
top5_m
```

```python
labels = ["กลิ่น","จราจร","ถนน","ทางเดิน","น้ำท่วม","มืด","สวน","สัตว์","สุขภาพกาย","จิต","ไม่ปลอดภัย"]
all_m=res_all.iloc[:,1].astype(int)#ปัญหารวม
top5_m=res.iloc[:,1].astype(int)#ปัญหา เขต top_5_parks

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig,ax = plt.subplots()
rects1 = ax.bar(x - width/2,all_m, width, label='all')
rects2 = ax.bar(x + width/2,top5_m, width, label='top5_parks')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('%จำนวนการร้องเรียน')
ax.set_title('เทียบจำนวนการร้องเรียน')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
